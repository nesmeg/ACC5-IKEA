```python
import pandas as pd

import matplotlib.pyplot as plt

# Load the CSV file
df_visits = pd.read_csv('visits.csv')

# Convert the date column to datetime
df_visits['date_visit'] = pd.to_datetime(df_visits['date_visit'])

# Set the date column as the index
df_visits.set_index('date_visit', inplace=True)

#For the data in 'location' Barcelona, plot the number of visits per day in one plot, in another plot the average visits per weekday and in another plot the average visits per month.
df_visits_barcelona = df_visits[df_visits['location'] == 'Barcelona']
df_visits_madrid = df_visits[df_visits['location'] == 'Madrid']
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df_visits_barcelona['visits'])
plt.show()

plot_acf(df_visits_madrid['visits'])
plt.show()
```
